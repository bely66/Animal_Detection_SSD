## Preprocess File Names

```
import json
import os
path = "Boxing_KNPS_image/KNPS_Captures"


for file_name in os.listdir(path):
  values = file_name.split(".")
  format = values[-1]
  if format.find("jpg")!=-1:
    format = format.upper()
    values[-1] = format
    new_format = ".".join(values)
    print(new_format)

    old_name = os.path.join(path, file_name)
    new_name = os.path.join(path, new_format)
    os.rename(old_name, new_name)
```

## Visualize Images

```
label_map = {k:v for v, k in label_map.items()}
```

```
import cv2


font                   = cv2.FONT_HERSHEY_SIMPLEX
bottomLeftCornerOfText = (10,500)
fontScale              = 5
fontColor              = (255,0,0)
lineType               = 14
image_list = []
label_list = []
for index, img_path in enumerate(data):
  img = cv2.imread(img_path)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  cords = labels[index]["boxes"]
  img_labels = labels[index]["labels"]
  for b_index, box in enumerate(cords):
    cv2.rectangle(img,tuple(box[:2]),tuple(box[2:]),(0,255,0),2)
    
    obj_name = label_map[img_labels[b_index]]
    cv2.putText(img,obj_name, 
    (box[3], box[1]), 
    font, 
    fontScale,
    fontColor,
    lineType)
  label_list.append(img_labels)
  image_list.append(img)
```

```
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import norm

def brightness(img):
    if len(img.shape) == 3:
        # Colored RGB or BGR (*Do Not* use HSV images with this function)
        # create brightness with euclidean norm
        return np.average(norm(img, axis=2)) / np.sqrt(3)
    else:
        # Grayscale
        return np.average(img)
pic = image_list[-10]
plt.imshow(pic)
print(brightness(pic))
```

```
plt.figure(figsize=(14,14)) # specifying the overall grid size

for i in range(16):
    plt.subplot(4,4,i+1)    # the number of images in the grid is 5*5 (25)
    plt.imshow(image_list[i])
    title = ""
    for img_label in label_list[i]:
      title += " " + label_map[img_label]
    plt.title(title)

plt.show()
```
